### Number of taxons in each cluster:

```python
directory = 'group/genes'

table['Thermoanaerobacter'] = 0
table['Aerococcus'] = 0
table['Anoxybacillus'] = 0
table['Paenibacillus'] = 0
table['Enterococcus'] = 0
table['virgibacillius'] = 0
table['Streptococcus'] = 0
table['Staphylococcus'] = 0

for root, dirs, files in os.walk(directory):
    for file in files:
        
        name = os.path.join(root, file)
        genes = pd.array(pd.read_csv(name, names=['gene'])['gene'])
        species = file[file.find('_')+1:file.find('.')]
        taxon = file[:file.find('_')]
        
        #  number of genes in a taxon
        for i in range(len(genes)):
            table[taxon] += pd.Series(table[taxon+'_'+species+'.faa'].str.contains(genes[i], regex=False)) 
        #table.loc[table[taxon+'_'+species+'.faa'].isin(genes), taxon] += 1
        #print(genes)
        #break
```

```python
genes = np.zeros(len(table), dtype=int)

for i in range(len(table)):
    genes[i] = ''.join(pd.array(table.iloc[i])[3:].astype(str)).count('.')
```

```python
taxons = np.zeros(len(table), dtype=int)

for i in range(len(table)):
    tmp = ''.join(pd.array(table.iloc[i])[3:8].astype(str)).count('*') != 5
    tmp += ''.join(pd.array(table.iloc[i])[8:13].astype(str)).count('*') != 5
    tmp += ''.join(pd.array(table.iloc[i])[13:18].astype(str)).count('*') != 5
    tmp += ''.join(pd.array(table.iloc[i])[18:23].astype(str)).count('*') != 5
    tmp += ''.join(pd.array(table.iloc[i])[23:28].astype(str)).count('*') != 5
    tmp += ''.join(pd.array(table.iloc[i])[28:33].astype(str)).count('*') != 5
    tmp += ''.join(pd.array(table.iloc[i])[33:38].astype(str)).count('*') != 5
    tmp += ''.join(pd.array(table.iloc[i])[38:43].astype(str)).count('*') != 5
    taxons[i] = tmp

np.save('all_taxons', taxons)
```
